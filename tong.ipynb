```
n = int(input("Nhập số lượng phần tử n: "))
z = []
for x in range(n):
    y =int(input(f"Nhập phần tử thứ {x+1}: "))
    z.append(y)

tong_chan = 0
tong_le = 0
for x in range(n):
    if z[x] % 2 == 0:
        tong_chan = tong_chan + z[x]
    else:
        tong_le = tong_le + z[x]
print("Tong chan =", tong_chan)
print("Tong le =", tong_le)
```
